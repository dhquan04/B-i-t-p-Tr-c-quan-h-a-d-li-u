# Superstore

```python
import numpy as np
import pandas as pd
import os
```

```python
data = pd.read_csv('Superstore.csv')
data
```

```python
data.shape
```

```python
types=np.unique(data.dtypes)
print(types)
```

```python
# Phân loại cột theo kiểu định lượng và định tính
quantitative = []  # Cột định lượng (numerical)
qualitative = []   # Cột định tính (categorical)

for column in data.columns:
    if pd.api.types.is_object_dtype(data[column]):
        quantitative.append(column)
    else:
        qualitative.append(column)

print("\nCác cột định lượng :", quantitative)
print("Các cột định tính :", qualitative)
```

```python
for i in range(len(types)):
    if (i ==0):
        print(' int64')
    elif(i == 1 ):
        print('float64')
    elif(i==2):
        print('object')
    
    for col in data.columns:
        if data[col].dtypes==types[i]:
            print("-------", col)
    
```

```python
#  In các cột có dữ liệu bị mất
missing_columns = data.columns[data.isnull().any()]
print("\nCác cột có dữ liệu bị mất:")
print(missing_columns)

#  Xử lý dữ liệu bị mất theo các phương pháp khác nhau

#  Loại bỏ các hàng có giá trị bị thiếu
data_dropna = data.dropna()
print("\nKích thước dữ liệu sau khi loại bỏ các hàng có giá trị bị thiếu:", data_dropna.shape)

# Thay thế giá trị bị thiếu bằng giá trị trung bình (nếu là kiểu số)
data_fillna_mean = data.copy()
for col in missing_columns:
    if data[col].dtype in ['float64', 'int64']:
        data_fillna_mean[col] = data[col].fillna(data[col].mean())
print("\nKích thước dữ liệu sau khi thay thế giá trị bị thiếu bằng giá trị trung bình:", data_fillna_mean.shape)

#  Thay thế giá trị bị thiếu bằng giá trị thường xuyên nhất (mode)
data_fillna_mode = data.copy()
for col in missing_columns:
    data_fillna_mode[col] = data[col].fillna(data[col].mode()[0])
print("\nKích thước dữ liệu sau khi thay thế giá trị bị thiếu bằng giá trị thường xuyên nhất:", data_fillna_mode.shape)
```
